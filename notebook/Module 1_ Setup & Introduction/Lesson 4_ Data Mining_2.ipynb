#  Classification: Basic Concepts and Techniques

## General idea behind classification.

The data for a classification task consists of a collection of instances (records).

Each such instance is characterized by the tuple **(x , y)**, where x is the set of attribute values that describe the instance and y is the class label of the instance.

The attribute set can contain attributes of any type, while the class **label y must be categorical**.

# Association Rule Mining Basics

## What is an **Itemset**?

An **itemset** is a collection of one or more items that appear together in a transaction or dataset record.

### Example:

Consider a dataset of shopping transactions:

| Transaction ID | Items                     |
|----------------|---------------------------|
| T1             | Bread, Cheese, Milk       |
| T2             | Bread, Milk               |
| T3             | Cheese, Milk              |

Some possible itemsets are:

- `{Milk}`
- `{Bread, Milk}`
- `{Cheese, Milk}`

Itemsets can be of any size:  
- 1-itemset (single item)  
- 2-itemset (pairs of items)  
- 3-itemset, and so on.

---

## What is **Support Count**?

The **support count** of an itemset is the number of transactions in which the itemset appears.

### Example:

From the transactions above:

| Itemset         | Support Count (Number of Transactions Containing Itemset) |
|-----------------|------------------------------------------------------------|
| `{Milk}`        | 3 (appears in T1, T2, T3)                                  |
| `{Bread, Milk}` | 2 (appears in T1, T2)                                      |
| `{Cheese, Milk}`| 2 (appears in T1, T3)                                      |

---

## What is **Support**?

Support is the relative frequency of an itemset in the dataset, calculated as:
image.png
\[
\text{Support}(X) = \frac{\text{Support Count}(X)}{\text{Total Number of Transactions}}
\]

### Example:

- Support of `{Milk}` = 3 / 3 = 1.0 (100%)  
- Support of `{Bread, Milk}` = 2 / 3  0.67 (67%)

---

## Summary

| Term            | Definition                                       |
|-----------------|-------------------------------------------------|
| **Itemset**     | A set of items occurring together in transactions |
| **Support Count** | The number of transactions containing the itemset |
| **Support**     | The fraction of total transactions containing the itemset |

### Association Rule

# What is an **Association Rule**?

An **Association Rule** describes a relationship between items in a dataset, typically transactions. It is a statement of the form:

> **If** a transaction contains itemset **A**,  
> **then** it is likely to also contain itemset **B**.

This is usually written as:

\[
A -> B
\]

where \(A\) and \(B\) are disjoint itemsets (they have no items in common).


### Example:

In a supermarket dataset:

\[
{Bread, Butter} -> {Jam}
\]

Means: If a customer buys bread and butter, they are likely to also buy jam.

---

## Metrics to Evaluate Association Rules

1. **Support**

- Measures how frequently the itemsets appear together in the dataset.


2. **Confidence**

- Measures how often items in \(B\) appear in transactions that contain \(A\).


image.png

## Why Use Support and Confidence?

Support is an important measure because a rule that has very low support
might occur simply by chance.
