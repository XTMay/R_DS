In R, the double colon :: is used to explicitly call a function from a specific package, even if that package isn‚Äôt attached with library().


```
So when you write:

‚ÄúUse the skim() function from the skimr package.‚Äù

This is useful when:

	‚Ä¢	You want to avoid conflicts between functions of the same name from different packages.
	‚Ä¢	You want to use a function without loading the entire package with library(skimr).
```

````r
# World Bank filtering & reshaping

library(dplyr) # Explicitly load dplyr for select and other functions
library(tidyr) # Explicitly load tidyr for pivot_longer
library(stringr) # Explicitly load stringr for str_c

wb_filt <- wb_data |>
  filter(`Country Name` %in% countries) |> #  Filters the dataset to only include rows where Country Name is in a predefined list countries.
  pivot_longer(cols = `2010`:`2023`, # Reshapes the data from wide to long format: ‚Ä¢	Year columns (2010‚Äì2023) become a single Year column. ‚Ä¢	Corresponding values go into a new column EduExp_GDP (education expenditure as % of GDP).
  names_to = "Year",
  values_to = "EduExp_GDP") |>
  mutate(Key = str_c(`Country Name`, Year)) |>
  select(`Country Name`, Year, Key, EduExp_GDP) |>
  filter(!is.na(EduExp_GDP)) # Na remving missing value
````

#### In R, Exploratory Data Analysis (EDA) is the process of summarizing and visualizing the main characteristics of a dataset. Here are common approaches:

#### üßπ 1. Data Cleaning

	‚Ä¢	janitor::clean_names() ‚Äì simplify column names
	‚Ä¢	dplyr::filter(), mutate(), select() ‚Äì data wrangling
	‚Ä¢	tidyr::drop_na(), replace_na() ‚Äì handle missing values


#### üìä 2. Summary Statistics

	‚Ä¢	summary() ‚Äì basic summaries
	‚Ä¢	skimr::skim() ‚Äì enhanced summaries (e.g., histograms, missing values)
	‚Ä¢	psych::describe() ‚Äì detailed statistics


#### üìà 3. Visualization
	‚Ä¢	ggplot2::ggplot() ‚Äì flexible grammar of graphics
	‚Ä¢	GGally::ggpairs() ‚Äì pairwise plots
	‚Ä¢	boxplot(), hist(), plot() ‚Äì base R visuals

#### üîç 4. Variable Exploration
	‚Ä¢	Numeric: mean(), sd(), cor(), ggplot2 histograms or scatterplots
	‚Ä¢	Categorical: table(), prop.table(), bar plots with ggplot2


#### üß† 5. Multivariate Analysis (Optional)
	‚Ä¢	Correlation matrices
	‚Ä¢	PCA: prcomp()
	‚Ä¢	Clustering: kmeans(), hclust()

#### üî† Example Packages for EDA
	‚Ä¢	tidyverse: core for data handling
	‚Ä¢	skimr: summary
	‚Ä¢	GGally: pairwise plots
	‚Ä¢	DataExplorer, summarytools: full EDA reports
