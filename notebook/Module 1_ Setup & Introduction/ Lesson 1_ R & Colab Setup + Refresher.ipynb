In R, the double colon :: is used to explicitly call a function from a specific package, even if that package isn’t attached with library().


```
So when you write:

skimr::skim(dataset)

It means:

“Use the skim() function from the skimr package.”

This is useful when:

	•	You want to avoid conflicts between functions of the same name from different packages.
	•	You want to use a function without loading the entire package with library(skimr).

```



✅ Equivalent to:

````r
# World Bank filtering & reshaping

library(dplyr) # Explicitly load dplyr for select and other functions
library(tidyr) # Explicitly load tidyr for pivot_longer
library(stringr) # Explicitly load stringr for str_c

wb_filt <- wb_data |>
  filter(`Country Name` %in% countries) |> #  Filters the dataset to only include rows where Country Name is in a predefined list countries.
  pivot_longer(cols = `2010`:`2023`, # Reshapes the data from wide to long format: •	Year columns (2010–2023) become a single Year column. •	Corresponding values go into a new column EduExp_GDP (education expenditure as % of GDP).
  names_to = "Year",
  values_to = "EduExp_GDP") |>
  mutate(Key = str_c(`Country Name`, Year)) |>
  select(`Country Name`, Year, Key, EduExp_GDP) |>
  filter(!is.na(EduExp_GDP)) # Na remving missing value
````

#### In R, Exploratory Data Analysis (EDA) is the process of summarizing and visualizing the main characteristics of a dataset. Here are common approaches:

#### 🧹 1. Data Cleaning

	•	janitor::clean_names() – simplify column names
	•	dplyr::filter(), mutate(), select() – data wrangling
	•	tidyr::drop_na(), replace_na() – handle missing values


#### 📊 2. Summary Statistics

	•	summary() – basic summaries
	•	skimr::skim() – enhanced summaries (e.g., histograms, missing values)
	•	psych::describe() – detailed statistics


#### 📈 3. Visualization
	•	ggplot2::ggplot() – flexible grammar of graphics
	•	GGally::ggpairs() – pairwise plots
	•	boxplot(), hist(), plot() – base R visuals

#### 🔍 4. Variable Exploration
	•	Numeric: mean(), sd(), cor(), ggplot2 histograms or scatterplots
	•	Categorical: table(), prop.table(), bar plots with ggplot2


#### 🧠 5. Multivariate Analysis (Optional)
	•	Correlation matrices
	•	PCA: prcomp()
	•	Clustering: kmeans(), hclust()

#### 🔠 Example Packages for EDA
	•	tidyverse: core for data handling
	•	skimr: summary
	•	GGally: pairwise plots
	•	DataExplorer, summarytools: full EDA reports
