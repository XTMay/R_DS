## 1. Background & Motivation

- Refugee movements are shaped by:
  - **Push factors**: conflict, political repression, poverty, economic collapse.
  - **Pull factors**: income opportunities, welfare systems, asylum policies, diaspora networks.
- Refugees place short-term fiscal pressure on host governments, especially in **education**.
- Central research question:  
  *How do GDP and refugee numbers interact, and how does this interaction affect education expenditure as a share of GDP?*

## 2. GDP and Refugee Flows

### Origin Countries
- **Neumayer (2005, ISQ ‚ÄúBogus Refugees?‚Äù):**
  - Cross-national asylum applications data (1980s‚Äì1990s).
  - Found **origin GDP per capita ‚Üì ‚Üí asylum outflows ‚Üë**.
  - But political repression and conflict dominate economic predictors.
- **Hatton (2016, Economic Policy):**
  - Panel of OECD/EU asylum inflows, 1985‚Äì2010.
  - Confirms **negative link** between origin income levels and refugee outflows.
  - Adds that growth shocks (GDP decline) significantly drive refugee numbers.

 **Conclusion:** Lower GDP per capita at origin correlates with higher refugee outflows, though conflict remains the primary driver.

### Destination Countries
- **Neumayer (2005, EU Politics ‚ÄúAsylum Destination Choice‚Äù):**
  - Modeled EU asylum seeker destinations.
  - **Higher destination GDPpc and higher recognition rates ‚Üí more applications**.
  - Distance and diaspora networks are also strong predictors.
- **EU-focused study (2008‚Äì2020, World Development):**
  - First-time asylum applications to the EU.
  - Reaffirmed GDP and asylum policy as **consistent pull factors**.

  **Conclusion:** Destination GDP is a strong pull factor, but policies (recognition rates, asylum generosity) condition its effect.

## 3. Refugees and Host-Country GDP

- **d‚ÄôAlbis, Boubtane & Coulibaly (2018, Science Advances):**
  - Studied asylum inflows into Western Europe (1985‚Äì2015).
  - Found **positive effect of asylum inflows on GDP and employment**.
  - No negative effect on fiscal balance in the medium run.
- **IMF (2023 Working Paper):**
  - Expanded dataset to global migration episodes.
  - Used **instrumental variables & local projection methods**.
  - Found **modest but positive GDP growth effects** from refugee inflows, especially when labor markets are flexible.

 **Conclusion:** Refugee inflows are not just a fiscal cost; they can stimulate growth in the medium run.

## 4. Refugees and Education Expenditure (% of GDP)

###  Global Evidence
- **World Bank & UNHCR (2021, 2023 ‚Äì Global Costing Studies):**
  - Simulation of education costs if all refugee children enrolled in national systems.
  - Cost ‚âà **<5% of host education budgets** on average, though higher in low-income countries.
  - Strong emphasis on international support to manage variation.

###  Country Evidence
- **Jordan Case Study (World Bank, 2016‚Äì2019):**
  - Syrian refugee influx ‚Üí increased student numbers.
  - Documented **budget shortfalls** and higher spending needs compared to peers at similar income levels.
  - Reliance on donor aid to stabilize budgets.

###  Related Migration Studies
- General immigration studies (e.g., OECD contexts):
  - Small **negative effect on per-pupil spending** when migrant numbers rise, due to fiscal dilution.
  - But effect sizes are context-specific.

 **Conclusion:** Refugee inflows increase education expenditure needs, but relative to GDP the impact is usually modest when donor aid is considered.
