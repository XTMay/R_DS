````r
# ==============================
# Load required libraries
# ==============================
library(dplyr)
library(ggplot2)
library(tidyr)
library(corrplot)
library(caret)
library(randomForest)
library(broom)
library(scales)
````

````r
# ==============================
# 1. Data Preprocessing
# ==============================
df <- df_gdp_refugees

# Convert numeric columns
num_cols <- c("edu_exp_gdp", "gdp_per_capita", "gov_exp_gdp", "unemployment_rate",
              "inflation_rate", "population", "urban_pop_percent", "education_index",
              "hdi_proxy", "Refugees", "Returned.Refugees", "Asylum.Seekers",
              "IDPs", "Returned.IDPs", "Stateless", "HST", "OOC")

df[num_cols] <- lapply(df[num_cols], function(x) as.numeric(as.character(x)))

# Convert logical columns to factor # bool, T/F
df$Is_Origin <- as.factor(df$Is_Origin)
df$Is_Destination <- as.factor(df$Is_Destination)
df$Is_Mixed <- as.factor(df$Is_Mixed)

# Handle missing values: replace NAs with median/mean for numeric, "Unknown" for categorical
# Handle missing values: if there are many NA, NA > 50% , 61/100 -> remove it, fill in values (True)
# filling value ---> real_value
df[num_cols] <- lapply(df[num_cols], function(x) ifelse(is.na(x), median(x, na.rm=TRUE), x))
df$region[is.na(df$region)] <- "Unknown"
````

````r
# ==============================
# 2. Feature Engineering
# ==============================
df <- df %>%
  mutate(
    Refugee_Return_Rate = Returned.Refugees / ifelse(Refugees == 0, 1, Refugees),
    IDP_Return_Rate = Returned.IDPs / ifelse(IDPs == 0, 1, IDPs),
    Edu_per_Capita = edu_exp_gdp * gdp_per_capita,
    Urban_Population = population * urban_pop_percent / 100,
    # new_feature = formula
    Refugee_Flow_Category = case_when(
      Refugee_Return_Rate > 0.5 ~ "High_Return",
      Refugee_Return_Rate > 0.1 ~ "Medium_Return",
      TRUE ~ "Low_Return"
    )
  )

df$Refugee_Flow_Category <- as.factor(df$Refugee_Flow_Category)
````

````r
# ==============================
# 3. Exploratory Data Analysis (EDA)
# ==============================
summary(df)
str(df)

# Distribution plots
ggplot(df, aes(x=edu_exp_gdp)) + geom_histogram(binwidth=1, fill="steelblue", color="black") +
  labs(title="Distribution of Education Expenditure (% GDP)", x="Education Expenditure", y="Count")

ggplot(df, aes(x=Refugee_Return_Rate)) + geom_histogram(binwidth=0.05, fill="orange", color="black") +
  labs(title="Distribution of Refugee Return Rate", x="Return Rate", y="Count")

# Boxplot by Region
ggplot(df, aes(x=region, y=Refugee_Return_Rate)) +
  geom_boxplot(fill="lightgreen") +
  theme(axis.text.x = element_text(angle = 45, hjust = 1)) +
  labs(title="Refugee Return Rate by Region", x="Region", y="Return Rate")
````

````r
# ==============================
# 4. Correlation Analysis
# ==============================
num_data <- df %>%
  select(edu_exp_gdp, gdp_per_capita, gov_exp_gdp, unemployment_rate,
         inflation_rate, population, urban_pop_percent, education_index,
         hdi_proxy, Refugees, Returned.Refugees, IDPs, Returned.IDPs)

cor_matrix <- cor(num_data)
corrplot(cor_matrix, method="number", type="upper", tl.cex=0.8)
````

````r
# ==============================
# 5. Linear Regression
# Predict Refugee Return Rate
# ==============================
lm_model <- lm(Refugee_Return_Rate(y) ~ edu_exp_gdp + gdp_per_capita + education_index + hdi_proxy + unemployment_rate + inflation_rate (x1 .. x7 ), data=df)
summary(lm_model)

# Visualize predictions
df$predicted_return <- predict(lm_model, df)
ggplot(df, aes(x=predicted_return, y=Refugee_Return_Rate)) +
  geom_point(color="blue") + geom_abline(intercept=0, slope=1, color="red") +
  labs(title="Predicted vs Actual Refugee Return Rate", x="Predicted", y="Actual")
````

````r
# ==============================
# 6. Classification for Policy Insight
# ==============================
# Target: Refugee_Flow_Category (High/Medium/Low return)
set.seed(123)
trainIndex <- createDataPartition(df$Refugee_Flow_Category, p = .8, list = FALSE)
train <- df[trainIndex, ]
test  <- df[-trainIndex, ]

# Random Forest Classifier
rf_model <- randomForest(Refugee_Flow_Category ~ edu_exp_gdp + gdp_per_capita + education_index + hdi_proxy + unemployment_rate + inflation_rate,
                         data=train, importance=TRUE, ntree=200)

# Predictions
pred <- predict(rf_model, test)
confusionMatrix(pred, test$Refugee_Flow_Category)

# Feature importance
importance(rf_model)
varImpPlot(rf_model)
````

# Research Report: Determinants of Refugee Flows and Policy Insights
## 研究報告：難民流出の要因と政策的示唆

---

## 1. Executive Summary / 要約

**English:**  
Forced displacement and refugee movements are shaped by a complex interaction of push factors (conflict, persecution, economic collapse, climate disasters) and pull factors (safety, policies, networks, opportunities). By monitoring quantifiable indicators, governments and international organizations can predict potential refugee surges and design evidence-based responses.

**日本語:**  
強制移住や難民の移動は、紛争・迫害・経済崩壊・気候災害といった「プッシュ要因」と、安全性・政策・社会ネットワーク・経済的機会といった「プル要因」の複雑な相互作用によって決定されます。定量的な指標を監視することで、各国政府や国際機関は難民流入の兆候を予測し、根拠に基づいた対応策を設計できます。

---

## 2. Key Drivers of Refugee Outflows / 難民流出の主な要因

| Category / カテゴリ | Indicators / 指標 | Sources / データソース | Expected Impact / 影響方向 |
|---------------------|------------------|-----------------------|---------------------------|
| Conflict & Violence / 紛争・暴力 | Battle deaths, attacks, conflict events / 戦闘死亡者数・紛争事件数 | UCDP, ACLED | ↑ outflows / 流出増加 |
| Political Persecution / 政治的迫害 | Freedom House scores, repression reports / 自由度指標、弾圧報告 | UNHCR, Freedom House | ↑ outflows / 流出増加 |
| Economic Collapse / 経済崩壊 | GDP decline, unemployment, food price index / GDP減少、失業率、食料価格指数 | World Bank, FAO, IMF | ↑ outflows / 流出増加 |
| Climate & Disasters / 気候・災害 | Drought index, flood victims / 干ばつ指数、洪水被災者数 | EM-DAT, UNEP | ↑ displacement / 流出増加 |
| Social Networks / 社会ネットワーク | Diaspora size, ethnic/language ties / ディアスポラ規模、民族・言語的つながり | Migration Data Portal, UNHCR | ↑ destination pull / 目的国吸引力 |
| Policy & Accessibility / 政策とアクセス | Asylum approval rates, border openness / 庇護承認率、国境の開放度 | UNHCR, IOM, OECD | ↑ flexible policy → inflows / 柔軟政策→流入増加 |

## 3. Actionable Policy Insights / 政策的インサイト

**English:**  
1. **Early Warning Systems**: Monitor spikes in violence, food prices, and climate disasters.  
2. **Regional Burden Sharing**: Strengthen cross-country mechanisms for temporary protection.  
3. **Adaptive Border Policies**: Allow controlled, humanitarian entry during crises.  
4. **Invest in Resilience**: Support vulnerable communities in climate and economic recovery.  
5. **Transparent Communication**: Share risk maps and forecasts to reduce panic and strengthen public trust.  

**日本語:**  
1. **早期警戒システム**：暴力、食料価格、気候災害の急増を「トリガー」として監視。  
2. **地域的な負担分担**：一時的保護のための越境メカニズムを強化。  
3. **適応的な国境政策**：危機時には人道的入国を認める仕組みを整備。  
4. **レジリエンス投資**：気候や経済から影響を受けやすい地域社会を支援。  
5. **透明なコミュニケーション**：リスクマップや予測を公表し、混乱を抑え社会的信頼を強化。
