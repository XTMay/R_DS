## Data preparation

### Subtask:
Clean column names, add a group column, combine dataframes, and convert date and salary columns to appropriate types.


**Reasoning**:
Install and load the necessary R libraries, create a list of dataframes, and loop through them to add a group column, clean names, rename columns, convert date columns, and convert salary columns to numeric, then combine into a single dataframe.



```r
# Install and load necessary libraries if not already installed
if (!requireNamespace("readxl", quietly = TRUE)) install.packages("readxl")
if (!requireNamespace("dplyr", quietly = TRUE)) install.packages("dplyr")
if (!requireNamespace("lubridate", quietly = TRUE)) install.packages("lubridate")
if (!requireNamespace("janitor", quietly = TRUE)) install.packages("janitor")
if (!requireNamespace("stringr", quietly = TRUE)) install.packages("stringr")

library(readxl)
library(dplyr)
library(lubridate)
library(janitor)
library(stringr)

# Create a list of the dataframes and their corresponding group names
dataframe_list <- list(
  "Saudi Arabia" = all_Saudi_Arabia,
  "Japanese" = all_Japanese,
  "Other" = all_other
)

# Initialize an empty list to store processed dataframes
processed_dataframes <- list()

# Loop through the list of dataframes
for (group_name in names(dataframe_list)) {
  df <- dataframe_list[[group_name]]

  # Add a new column named group
  df <- df %>%
    mutate(group = group_name)

  # Clean the column names
  df <- df %>%
    clean_names()

  # Select and rename the relevant columns
  # Based on previous inspection, the relevant columns are:
  # ser_no, emp_no, employee_name, date_of_joining, date_of_leaving, dept, column7, column8, column9, column10, column11, column12
  # Rename them to more descriptive names
  df <- df %>%
    select(
      ser_no,
      emp_no,
      employee_name,
      date_of_joining,
      date_of_leaving,
      occupation = dept,
      nationality = column7,
      basic_salary = column8,
      housing_allowance = column9,
      transportation_allowance = column10,
      food_allowance = column11,
      total_salary = column12,
      group
    )

  # Convert date_of_joining and date_of_leaving to Date
  # Since they are already POSIXct, direct conversion to Date should work
  df <- df %>%
    mutate(
      date_of_joining = as_date(date_of_joining),
      date_of_leaving = as_date(date_of_leaving)
    )

  # Convert salary-related columns to numeric, removing commas and coercing errors to NA
  salary_cols <- c("basic_salary", "housing_allowance", "transportation_allowance", "food_allowance", "total_salary")

  for (col in salary_cols) {
    df <- df %>%
      mutate(!!col := as.numeric(str_replace_all(!!sym(col), ",", ""))) # Use !! and sym() for NSE
  }

  # Store the processed dataframe in the list
  processed_dataframes[[group_name]] <- df
}

# Combine the processed dataframes into a single dataframe
all_data <- bind_rows(processed_dataframes)

# Display the structure and the first few rows of the combined dataframe
print(str(all_data))
head(all_data)
```

## Feature engineering

### Subtask:
Create new variables for tenure, attrition status, and year components, and handle missing/inconsistent data.


**Reasoning**:
Calculate tenure, create attrition status, extract year components, and handle potential inconsistencies.



```r
# Calculate tenure_days
all_data <- all_data %>%
  mutate(tenure_days = as.numeric(date_of_leaving - date_of_joining)) # Calculate difference in days

# Create is_left variable
all_data <- all_data %>%
  mutate(is_left = if_else(!is.na(date_of_leaving), 1, 0))

# Extract year_joined and year_left
all_data <- all_data %>%
  mutate(
    year_joined = year(date_of_joining),
    year_left = year(date_of_leaving)
  )

# Handle inconsistent tenure_days (negative values)
# Negative tenure days likely indicate data entry errors or inconsistent date formats
# Convert negative tenure_days to NA
all_data <- all_data %>%
  mutate(tenure_days = if_else(tenure_days < 0, as.numeric(NA), tenure_days))

# Display the structure and first few rows with the new columns
print(str(all_data))
head(all_data)
```
