## ✅ 1. データの読み込み | Data Import (Raw)

````r
# CSV ファイルの読み込み / Read CSV file
data <- read.csv("data.csv")

# Excel ファイルの読み込み / Read Excel file
# install.packages("readxl")
library(readxl)
data <- read_excel("data.xlsx")
````

## ✅ 2. データの前処理 | Data Cleaning & Preprocessing

````r
# install.packages("dplyr")
library(dplyr)

# 欠損値の除去 / Remove NA
clean_data <- na.omit(data)

# 特定の列を抽出 / Select columns
data %>% select(name, score)

# フィルター / Filtering rows
data %>% filter(score > 80)

# 新しい列を追加 / Add calculated column
data %>% mutate(score2 = score * 1.1)
````

## ✅ 3. 要約統計量 | Summary Statistics

````r
summary(data)         # 基本統計量 / Basic summary
mean(data$score)      # 平均 / Mean
sd(data$score)        # 標準偏差 / Standard Deviation
table(data$gender)    # カテゴリ数 / Frequency table
````

## ✅ 4. データ可視化 | Data Visualization

````r
# install.packages("ggplot2")
library(ggplot2)

# 散布図 / Scatter Plot
ggplot(data, aes(x=age, y=score)) +
  geom_point()

# 箱ひげ図 / Boxplot
ggplot(data, aes(x=gender, y=score)) +
  geom_boxplot()

# ヒストグラム / Histogram
ggplot(data, aes(x=score)) +
  geom_histogram(binwidth=5)
````

## ✅ 5. 相関と回帰分析 | Correlation & Regression

image.png

````r
cor(data$age, data$score)  # 相関係数 / Correlation

# 単回帰分析 / Linear Regression
model <- lm(score ~ age, data = data)
summary(model)

# 散布図と回帰線 / Scatter + Regression Line
ggplot(data, aes(x=age, y=score)) +
  geom_point() +
  geom_smooth(method="lm")
````

## ✅ 6. クラスタリング / 分類 | Clustering / Classification

image.png

````r
# K-means クラスタリング
kmeans_result <- kmeans(data[, c("age", "score")], centers=3)
data$cluster <- kmeans_result$cluster

# 分類モデル例（ランダムフォレスト）
# install.packages("randomForest")
library(randomForest)
model <- randomForest(gender ~ age + score, data = data)
````

## ✅ 7. 時系列分析 | Time Series Analysis

image.png

````r
# 時系列データの作成
ts_data <- ts(data$sales, start=c(2020,1), frequency=12)

# 時系列プロット
plot(ts_data)

# ARIMA モデル
# install.packages("forecast")
library(forecast)
fit <- auto.arima(ts_data)
forecast(fit, h=6)
````

## 📊 回帰分析の主要統計量 / Key Regression Statistics

| 項目 / Item                        | 日本語の意味 / Meaning in Japanese                                                       | 英語の意味 / Meaning in English                                                  |
|-----------------------------------|------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------|
| 📉 p値 / p-value                   | 観測された結果が偶然に起こる確率。小さいほど統計的に有意（例：p < 0.05）。                        | Probability that the observed result occurred by chance. Smaller = more significant. |
| 📈 決定係数 / R²                   | 説明変数が目的変数のどれだけを説明できるか（0〜1）。1に近いほどモデルの説明力が高い。             | Proportion of variance in the dependent variable explained by the model (0–1).  |
| 🔍 調整済みR² / Adjusted R²        | 決定係数を変数の数で調整した指標。変数が多くても過剰適合しないように調整されている。               | R² adjusted for the number of predictors. Helps avoid overfitting.              |
| 🔁 残差標準誤差 / Residual SE      | モデルの予測誤差の平均的な大きさ。小さいほど予測精度が良い。                                     | Average size of the prediction errors. Smaller = better model fit.              |
| 📊 F統計量 / F-statistic           | モデル全体の有意性を示す。値が大きくp値が小さいほど、モデル全体が統計的に有意であることを示す。     | Indicates if the overall regression model is statistically significant.         |
