## âœ… 1. ãƒ‡ãƒ¼ã‚¿ã®èª­ã¿è¾¼ã¿ | Data Import (Raw)

````r
# CSV ãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿ / Read CSV file
data <- read.csv("data.csv")

# Excel ãƒ•ã‚¡ã‚¤ãƒ«ã®èª­ã¿è¾¼ã¿ / Read Excel file
# install.packages("readxl")
library(readxl)
data <- read_excel("data.xlsx")
````

## âœ… 2. ãƒ‡ãƒ¼ã‚¿ã®å‰å‡¦ç† | Data Cleaning & Preprocessing

````r
# install.packages("dplyr")
library(dplyr)

# æ¬ æå€¤ã®é™¤å» / Remove NA
clean_data <- na.omit(data)

# ç‰¹å®šã®åˆ—ã‚’æŠ½å‡º / Select columns
data %>% select(name, score)

# ãƒ•ã‚£ãƒ«ã‚¿ãƒ¼ / Filtering rows
data %>% filter(score > 80)

# æ–°ã—ã„åˆ—ã‚’è¿½åŠ  / Add calculated column
data %>% mutate(score2 = score * 1.1)
````

## âœ… 3. è¦ç´„çµ±è¨ˆé‡ | Summary Statistics

````r
summary(data)         # åŸºæœ¬çµ±è¨ˆé‡ / Basic summary
mean(data$score)      # å¹³å‡ / Mean
sd(data$score)        # æ¨™æº–åå·® / Standard Deviation
table(data$gender)    # ã‚«ãƒ†ã‚´ãƒªæ•° / Frequency table
````

## âœ… 4. ãƒ‡ãƒ¼ã‚¿å¯è¦–åŒ– | Data Visualization

````r
# install.packages("ggplot2")
library(ggplot2)

# æ•£å¸ƒå›³ / Scatter Plot
ggplot(data, aes(x=age, y=score)) +
  geom_point()

# ç®±ã²ã’å›³ / Boxplot
ggplot(data, aes(x=gender, y=score)) +
  geom_boxplot()

# ãƒ’ã‚¹ãƒˆã‚°ãƒ©ãƒ  / Histogram
ggplot(data, aes(x=score)) +
  geom_histogram(binwidth=5)
````

## âœ… 5. ç›¸é–¢ã¨å›å¸°åˆ†æ | Correlation & Regression

image.png

````r
cor(data$age, data$score)  # ç›¸é–¢ä¿‚æ•° / Correlation

# å˜å›å¸°åˆ†æ / Linear Regression
model <- lm(score ~ age, data = data)
summary(model)

# æ•£å¸ƒå›³ã¨å›å¸°ç·š / Scatter + Regression Line
ggplot(data, aes(x=age, y=score)) +
  geom_point() +
  geom_smooth(method="lm")
````

## âœ… 6. ã‚¯ãƒ©ã‚¹ã‚¿ãƒªãƒ³ã‚° / åˆ†é¡ | Clustering / Classification

image.png

````r
# K-means ã‚¯ãƒ©ã‚¹ã‚¿ãƒªãƒ³ã‚°
kmeans_result <- kmeans(data[, c("age", "score")], centers=3)
data$cluster <- kmeans_result$cluster

# åˆ†é¡ãƒ¢ãƒ‡ãƒ«ä¾‹ï¼ˆãƒ©ãƒ³ãƒ€ãƒ ãƒ•ã‚©ãƒ¬ã‚¹ãƒˆï¼‰
# install.packages("randomForest")
library(randomForest)
model <- randomForest(gender ~ age + score, data = data)
````

## âœ… 7. æ™‚ç³»åˆ—åˆ†æ | Time Series Analysis

image.png

````r
# æ™‚ç³»åˆ—ãƒ‡ãƒ¼ã‚¿ã®ä½œæˆ
ts_data <- ts(data$sales, start=c(2020,1), frequency=12)

# æ™‚ç³»åˆ—ãƒ—ãƒ­ãƒƒãƒˆ
plot(ts_data)

# ARIMA ãƒ¢ãƒ‡ãƒ«
# install.packages("forecast")
library(forecast)
fit <- auto.arima(ts_data)
forecast(fit, h=6)
````
